```python
agg_df = df.groupby (["Age", 'Nationality', "LodgingRevenue", "BookingsCanceled"])["DaysSinceCreation"].mean().sort_values(ascending=False)
agg_df.head()
```

```python
agg_df = agg_df.reset_index()
agg_df.head()
```

```python
my_labels = ['0_18.0', '19_23', '24_30', '31_40', '41_70' ]
agg_df["AGE_CUT"] = pd.cut(x=agg_df["Age"], bins=[0, 18, 23, 30, 40, 70], labels=my_labels)
agg_df.tail(10)
```

```python
agg_df["costumer_level_based"] = [f"{i[0]}_{i[1]}_{i[2]}_{i[-1]}" for i in agg_df.values]
```

```python
agg_df = agg_df.loc[:, ["costumer_level_based", "DaysSinceCreation"]].groupby("costumer_level_based").agg({"DaysSinceCreation": "mean"}).sort_values(by="DaysSinceCreation", ascending=False).reset_index()
agg_df["costumer_level_based"].head()
```

```python
agg_df["SEGMENT"] = pd.qcut(agg_df["DaysSinceCreation"], 4, labels=["D", "C", "B", "A"])
agg_df.head()
```

```python
agg_df.groupby(["SEGMENT"]).agg({"DaysSinceCreation": ["mean", "max", "sum"]})
agg_df[agg_df["SEGMENT"] == "C"].describe()
```
