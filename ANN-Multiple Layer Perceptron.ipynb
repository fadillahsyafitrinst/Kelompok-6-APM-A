```python
#bagi dataset menjadi 2 segmen
# 1) training set untuk melatih dataset
# 2) set validasi untuk menguji keakuratan neural network

#ambil 380 data sampel pertama untuk dataset training
X_train = X_normalized[:,:380]
Y_train = Y[:,:380]

#sisanya sebanyak 189 data, dari data ke-381 menjadi set validasi
X_cv = X_normalized[:,381:]
Y_cv = Y[:,381:]
```

```python
#menggunakan fungsi sigmoid untuk output layer neural network
def sigmoid(z):
    s = 1/(1+np.exp(-z))
    return s
```

```python
#Defenisikan fungsi tanh(x) untuk hidden layer neural network
def tanh(z):
    s = (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))
    return s
```

```python
#definisikan fungsi 
def forward_prop(X,W1,W2,W3,b1,b2,b3):
    
    #First layer forward propogation
    Z1 = np.dot(W1,X)
    A1 = tanh(Z1 + b1)
    #Second layer forward propogation
    Z2 = np.dot(W2,A1)
    A2 = tanh(Z2 + b2)
    #Third layer forward propogation
    Z3 = np.dot(W3,A2)
    A3 = sigmoid(Z3 + b3) #menghasilkan vektor probabilitas 
    
    cache = {    
                  "Z1": Z1,
                  "A1": A1,
                  "Z2": Z2,
                  "A2": A2,
                  "Z3": Z3,
                  "A3": A3
            }
    return cache
```

```python
#Melakukan penurunan gradient dari neural network
#1) inisialisasi parameter bobot dan intersep secara acak
#2) jalankan forward propagation neural network
#3) hitung turunan dari bobot dan parameter intersep dengan back propagation
#4) sempurnakan parameter dengan turunan (3)
#5) ulangi langkah 1 - 4 

def gradient_descent(iterations,X,Y,alpha):
    
    #Random menginisialisasi parameter sebelum menjalankan algoritme
    W1 = np.random.randn(3,30)*0.01
    b1 = np.random.rand(3,1)
    W2 = np.random.randn(2,3)*0.01
    b2 = np.random.rand(2,1)
    W3 = np.random.rand(1,2)*0.01
    b3 = np.random.rand(1,1)
    dummy,m = X.shape
    
    caches = [] #menyimpan cost dari tiap iterasi
    count_vector = [] #menyimpan jumlah iterasi
    count = 0
    
    for i in range (1,iterations):
        
            count = count + 1
            
            count_vector.append(count)
        
            params = forward_prop(X,W1,W2,W3,b1,b2,b3) #forward propogation dengan parameters
            
            #tentukan nilai yang digunakan dalam back propagation dg kamus nilai yang dibuat dari forward propagation
            Z1 = params['Z1']
            Z2 = params['Z2']
            Z3 = params['Z3']
            A1 = params['A1']
            A2 = params['A2']
            A3 = params['A3']
            
            #Defenisikan fungsi biaya, tambahkan biaya setiap iterasi ke cache
            cost = -(1 / m)*np.sum(np.multiply(Y,np.log(A3)) + np.multiply((1-Y),np.log(1-A3)))
            caches.append(cost)
            
            #Back propogation untuk layer 3
            dA3 = -Y/A3 + (1-Y)/(1-A3)
            dZ3 = dA3 * sigmoid(Z3)*(1-sigmoid(Z3))
            dW3 = (1 / m)*np.dot(dZ3,A2.T)
            db3 = (1 / m)*np.sum(dZ3,axis=1,keepdims=True)
            
            #Back propogation untuk layer 2
            dA2 = np.dot(W3.T,dZ3)
            dZ2 = dA2*(1-np.power(tanh(Z2),2))
            dW2 = (1 / m)*np.dot(dZ2,A1.T)
            db2 = (1 / m)*np.sum(dZ2,axis=1,keepdims=True)
            
            #Back propogation untuk layer 1
            dA1 = np.dot(W2.T,dZ2)
            dZ1 = dA1*(1-np.power(tanh(Z1),2))
            dW1 = (1 / m)*np.dot(dZ1,X.T)
            db1 = (1 / m)*np.sum(dZ1,axis=1,keepdims=True)
            
            #tetapkan ulang parameter bobot dg turunan yang dihitung dari back propagation
            W1 = W1 - alpha*dW1
            W2 = W2 - alpha*dW2
            W3 = W3 - alpha*dW3
            
            #tetapkan ulang parameter bobot dg turunan yang dihitung dari back propagation
            b1 = b1 - alpha*db1
            b2 = b2 - alpha*db2
            b3 = b3 - alpha*db3
        
    return W1,W2,W3,b1,b2,b3,count_vector,caches
```

```python
#cek algoritma, kurva akan menurun dg iterasi, dan akhirnya mendatar
#membantu menentukan jumlah iterasi yang tepat untuk dijalankan dengan parameter yang sesuai
#catatan, learning rate yang digunakan 0,1

W1,W2,W3,b1,b2,b3,count,caches = gradient_descent(1000,X_cv,Y_cv,0.5)

plt.plot(count,caches,label='Cost')

plt.xlabel('Iteration')
plt.ylabel('Cost')

plt.title("Cost vs. Iteration")

plt.legend()

plt.show()
```

```python
def predict(X,Y,iterations,alpha,X_train,Y_train):

    W1,W2,W3,b1,b2,b3,count,caches = gradient_descent(iterations,X_train,Y_train,alpha)
    
    Z1 = np.dot(W1,X)
    A1 = tanh(Z1 + b1)
    Z2 = np.dot(W2,A1)
    A2 = tanh(Z2 + b2)
    Z3 = np.dot(W3,A2)
    A3 = sigmoid(Z3 + b3)
    
    dummy,m = A3.shape
    Y_prediction = np.zeros((1, m))
    
    for i in range(m):
        
        Y_prediction[0, i] = 1 if A3[0, i] > 0.5 else 0
        
    return Y_prediction
```

```python
#akurasi prediksi dari neural network, dibandingkan dengan training set dan cross validasi set
print("Train accuracy: {} %".format(100 - np.mean(np.abs(predict(X_train,Y_train,1000,0.5,X_train,Y_train) - Y_train)) * 100))
print("Cross validation accuracy: {} %".format(100 - np.mean(np.abs(predict(X_cv,Y_cv,1000,0.5,X_train,Y_train) - Y_cv)) * 100))
```

```python
#Memprediksi tumor ganas
dummy,m1 = X_train.shape
dummy,m2 = X_cv.shape

train_predict = predict(X_train,Y_train,1000,1.5,X_train,Y_train)
CV_predict = predict(X_cv,Y_cv,1000,1.5,X_train,Y_train)
count_true_pos = 0
count_train_pos = 0

count_true_pos_cv = 0
count_cv_pos = 0

for i in range (1,m1):
    if train_predict[0,i] == 1 and Y_train[0,i] == 1:
        count_true_pos = count_true_pos + 1
    if Y_train[0,i] == 1:
        count_train_pos = count_train_pos + 1
        
for i in range (1,m2):
    if CV_predict[0,i] == 1 and Y_cv[0,i] == 1:
        count_true_pos_cv = count_true_pos_cv + 1
    if Y_cv[0,i] == 1:
        count_cv_pos = count_cv_pos + 1

print("\nPOSITIVE PREDICTED AND TRUE POSITIVE\n")
        
print(str(count_true_pos) + " = positives predicted pada training set")
print(str(count_train_pos) + " = true positives pada training set")
print("Akurasi true positives pada training set : {} %".format(100-np.abs(100*((count_true_pos - count_train_pos)/count_train_pos))))

print("----------------------------------------------------------------")

print(str(count_true_pos_cv) + " = positives predicted pada cross validation set")
print(str(count_cv_pos) + " = true positives pada cross validation set")
print("Akurasi true positives pada cross validation set is: {} %".format(100-np.abs(100*((count_true_pos_cv - count_cv_pos)/count_true_pos_cv))))
```
